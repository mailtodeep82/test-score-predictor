Libraries

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
# configurations
data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 7, 8, 10],
    'Test_Score':    [55, 62, 68, 70, 78, 85, 90, 95]
}
TARGET_COL = "Test_Score" # target column name
```

```python
df = pd.DataFrame(data)
print("DataFrame shape:", df.shape)

#8 rows (samples/observations)
#2 columns (features/variables)
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# separate features and target
X = df.drop(columns=[TARGET_COL])
y = df[TARGET_COL]
```

```python
X.head()
```

```python
y.head()
```

```python
# 1. Create a new, blank "brain" (model) from our recipe.
model = LinearRegression()

# 2. "Train" the brain.
# This is the most important line! We're telling the model
# to "fit" itself to our X and y data. It's looking at all
# the dots and figuring out the best line.
model.fit(X, y)

print("Done! The model has been trained.")
```

```python
# Let's see what the model learned!

# This is the 'b' (the Intercept)
intercept = model.intercept_

# This is the 'm' (the Slope)
# It's in an "array" (a list), so we grab the first item [0]
slope = model.coef_[0]

print(f"Model's Intercept (b): {intercept:.2f}")
print(f"Model's Slope (m): {slope:.2f}")

print(f"\nThis means the model learned the formula:")
print(f"Score = ({slope:.2f} * Hours_Studied) + {intercept:.2f}")
```

```python
# Let's predict a score for someone who studied for 6 hours.
# (Note: We have to give it the 6 in double brackets [[6]])

hours_to_predict = pd.DataFrame({'Hours_Studied': [6]})
predicted_score = model.predict(hours_to_predict)

print(f"The model predicts a score of {predicted_score[0]:.2f} for someone who studied 6 hours.")

# You can even check the math!
# (Slope * 6) + Intercept
math_check = (slope * 6) + intercept
print(f"Our manual math check: {math_check:.2f} (It matches!)")
```

```python
import joblib

# Save the model to a file
joblib.dump(model, 'test_score_model.pkl')
```
